## Selecting Genre to keep at random

```python
import random

seed_value = 888
random.seed(seed_value)

def genreRandom(df):
    df["genre"] = df["genres"].apply(lambda x: random.choice(x.split(",")))
    df = df.drop("genres", axis=1)
    return df

df_movies = genreRandom(df_movies)
df_series = genreRandom(df_series)
```

```python
df_series["genre"].unique()
```

```python
def convertInt(df, col):
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df = df[df[col].notna()]
    df = df[df[col].astype(int) > 0 ]
    return df

df_movies = convertInt(df_movies, "runtimeMinutes")
df_movies = convertInt(df_movies, "startYear")
df_series = convertInt(df_series, "startYear")

df_movies = df_movies[df_movies["runtimeMinutes"]<180]
```
